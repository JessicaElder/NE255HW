```python
#importing packages
import random as r
import math as m
import matplotlib.pyplot as plt
import numpy as np
```

```python
#If point is inside circle, add to "inside" variable. Repeat for however many histories. Manipulate "inside" variable to get pi. 
inside = 0
histories = 100
for i in range(0, histories):
    x = r.random()**2
    y = r.random()**2
    if m.sqrt(x + y) < 1: 
        inside += 1
        
pi100 = (float(inside)/histories) * 4
print(pi100)
```

```python
inside = 0
histories = 1000
for i in range(0, histories):
    x = r.random()**2
    y = r.random()**2
    if m.sqrt(x + y) < 1: 
        inside += 1
        
pi1000 = (float(inside)/histories) * 4
print(pi1000)
```

```python
inside = 0
histories = 10000
for i in range(0, histories):
    x = r.random()**2
    y = r.random()**2
    if m.sqrt(x + y) < 1:
        inside += 1
        
pi10000 = (float(inside)/histories) * 4
print(pi10000)
```

```python
#Plotting experimental values as a function of histories.
pi = 3.14159
histories = [100, 1000, 10000]
experimental = [pi100, pi1000, pi10000]

plt.plot(histories, experimental)
plt.axhline(y = pi, color = 'r')
plt.xlabel('Histories')
plt.ylabel('Experimental Values')
plt.xscale('log')
plt.title('Monte Carlo Calculations of Pi')
plt.show()
```

```python
#Plotting absolute error and relative error as functions of number of histories. 
relerror100 = abs(pi100 - pi)
relerror1000 = abs(pi1000 - pi)
relerror10000 = abs(pi10000 - pi)
abserror100 = relerror100 / pi
abserror1000 = relerror1000 / pi
abserror10000 = relerror10000 / pi


relerror = [relerror100, relerror1000, relerror10000]
abserror = [abserror100, abserror1000, abserror10000]
histories = [100, 1000, 10000]

rel = plt.plot(histories, relerror, color = 'b', label = 'Relative Error')
ab = plt.plot(histories, abserror, color = 'm', label = 'Absolute Error')
plt.xlabel('Histories')
plt.ylabel('Error')  
plt.xscale('log')
plt.title('Relative and Absolute Errors From Calculations')
plt.legend()
plt.show()
```
